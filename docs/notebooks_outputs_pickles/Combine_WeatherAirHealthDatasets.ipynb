```python
weatherpath = '/Users/hrm/Documents/Dropbox/DSrelated/Metis/climate_related/WeatherData/'

weather = pd.read_csv(weatherpath+'DS3505_SurfaceDailyData_Newark2004-2014.csv',index_col=0)
prcp = pd.read_csv(weatherpath+'cdiac_WeatherData_2004-2014_precp.csv',index_col=0)
prcp = prcp.rename(columns = {'Year':'YR','Month':'MTH','Day':'DAY'})
```

```python
weather.head()
```

```python
prcp.head()
```

```python
# tmp0 = pd.merge(weather, prcp, on=['Date','YR','MTH','DAY'], how='outer')
tmp0 = pd.merge(weather, prcp, on=['YR','MTH','DAY'], how='outer')
tmp00 = tmp0.drop(['Date_y'], axis=1)
tmp00 = tmp00.rename(columns = {'Date_x':'Date'})
tmp00.head()
```

```python
airpath = '/Users/hrm/Documents/Dropbox/DSrelated/Metis/climate_related/AirQualityData/'
ozone = pd.read_csv(airpath+'meanDailyOzoneConc.csv',index_col=0)
ozone = ozone.rename(columns = {'date_str':'Date'})

pm25 = pd.read_csv(airpath+'meanDailyPM2-5Conc.csv',index_col=0)
pm25 = pm25.rename(columns = {'date_str':'Date'})
```

```python
ozone.head()
```

```python
pm25.head()
```

```python
tmp1 = pd.merge(ozone, pm25, on=['Date','YR','MTH','DAY'], how='outer')
tmp1.head()
```

```python
tmp2 = pd.merge(tmp00, tmp1, on=['Date','YR','MTH','DAY'], how='outer')
tmp2.dropna()
```

```python
healthpath = '/Users/hrm/Documents/Dropbox/DSrelated/Metis/climate_related/HeathData/'
EDvisits = pd.read_csv(healthpath+'NYCAsthmaEDVisits_summedCounts.csv',index_col=0)
```

```python
EDvisits.head()
```

```python
tmp3=pd.merge(tmp2, EDvisits, on=['Date','YR','MTH','DAY'], how='outer')
tmp3.dropna()
```
